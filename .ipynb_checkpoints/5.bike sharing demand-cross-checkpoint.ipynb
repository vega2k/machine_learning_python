```python
import pandas as pd
```

### Load Dataset

```python
train = pd.read_csv("data/bike/train.csv",parse_dates=['datetime'])
```

```python
print(train.shape)
print(train.columns)
print(train.info())
train.head()
#casual + registered = count
#train["datetime"].dt.year #month,day,hour,minute,second
```

```python
# 숫자의 높고 낮음을 맞추는 것은 Regression
# 예측한 값이 어떤 Category 에 속하느냐? Classification
```

```python
test = pd.read_csv("data/bike/test.csv",parse_dates=['datetime'])
print(test.shape)
print(test.columns)
print(test.info())
test.head()
```

```python
print('season ',train['season'].value_counts())   # 봄,여름,가을,겨울
print('holiday ',train['holiday'].unique())   # 공휴일
print('working ',train['workingday'].unique()) #1 근무일
print('weather ',train['weather'].unique())  #1이면 좋은 날씨
print('temp ',train['temp'].unique())     #온도
print('atemp ',train['atemp'].unique())   #체감온도
print('humidity ',train['humidity'].unique())  #습도
#casual + registered = count
#회원이 아님 + 회원임
```

```python
print(train['windspeed'].value_counts()) #풍속
```

```python
#datetime으로 부터 년월일시분초요일 컬럼 만들기
train['datetime-year'] = train['datetime'].dt.year
train['datetime-month'] = train['datetime'].dt.month
train['datetime-day'] = train['datetime'].dt.day
train['datetime-hour'] = train['datetime'].dt.hour
train['datetime-minute'] = train['datetime'].dt.minute
train['datetime-second'] = train['datetime'].dt.second
train['datetime-dayofweek']=train['datetime'].dt.dayofweek

test['datetime-year'] = test['datetime'].dt.year
test['datetime-month'] = test['datetime'].dt.month
test['datetime-day'] = test['datetime'].dt.day
test['datetime-hour'] = test['datetime'].dt.hour
test['datetime-minute'] = test['datetime'].dt.minute
test['datetime-second'] = test['datetime'].dt.second
test['datetime-dayofweek']=test['datetime'].dt.dayofweek
```

```python
print(train['datetime-year'].unique(), test['datetime-year'].unique())
print(train['datetime-month'].unique(), test['datetime-month'].unique())
print(train['datetime-day'].unique(), test['datetime-day'].unique())
print(train['datetime-hour'].unique(), test['datetime-hour'].unique())
print(train['datetime-minute'].unique(), test['datetime-minute'].unique())
print(train['datetime-second'].unique(), test['datetime-second'].unique())
print(train['datetime-dayofweek'].unique(), test['datetime-dayofweek'].unique())
```

```python
train.head()
```

```python
test.head()
```

```python
train['datetime-dayofweek'].unique()
```

```python
#one-hot encoding
train["datetime-monday"] = train["datetime-dayofweek"] == 0
train["datetime-tuesday"] = train["datetime-dayofweek"] == 1
train["datetime-wednesday"] = train["datetime-dayofweek"] == 2
train["datetime-thursday"] = train["datetime-dayofweek"] == 3
train["datetime-friday"] = train["datetime-dayofweek"] == 4
train["datetime-saturday"] = train["datetime-dayofweek"] == 5
train["datetime-sunday"] = train["datetime-dayofweek"] == 6
```

```python
train.tail()
```

```python
test["datetime-monday"] = test["datetime-dayofweek"] == 0
test["datetime-tuesday"] = test["datetime-dayofweek"] == 1
test["datetime-wednesday"] = test["datetime-dayofweek"] == 2
test["datetime-thursday"] = test["datetime-dayofweek"] == 3
test["datetime-friday"] = test["datetime-dayofweek"] == 4
test["datetime-saturday"] = test["datetime-dayofweek"] == 5
test["datetime-sunday"] = test["datetime-dayofweek"] == 6
```

## Train

```python
feature_names = ['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','datetime-year','datetime-hour','datetime-monday','datetime-tuesday','datetime-wednesday','datetime-thursday','datetime-friday','datetime-saturday','datetime-sunday']
```

```python
X_train = train[feature_names]

print(X_train.shape)
X_train.head()
```

```python
X_test = test[feature_names]

print(X_test.shape)
X_test.head()
```

```python
label_name = "count"

y_train = train[label_name]

print(y_train.shape)
y_train.head()
```

```python
# from sklearn.tree import DecisionTreeRegressor

# # random_state 랜덤하게 나오지 않도록 고정시켜 주는 것임
# model = DecisionTreeRegressor(random_state=37, max_depth=5)
# model
```

```python
#RandomForest 알고리즘 사용
# 66% 데이터를 사용
# 100개와 10개 사이에서 데이터를 맞출 가능성
#Sampling 해서 여래개의 트리를 계속해서 맞춰서 사용한다. 
#투표시스템을 사용한다.

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=37)
model
```

### Cross Validation

```python
from sklearn.model_selection import cross_val_score
cross_val_score(model,X_train,y_train,scoring="neg_mean_absolute_error",cv=20)  #default 가 3조각, cv=20,
```

```python
#decision tree Score = 76.10323
#randaom forest = 33.51522
score = cross_val_score(model,X_train,y_train,cv=20,scoring="neg_mean_absolute_error").mean()
print("Score = {0:.5f}".format(score))
score = -1.0 * score
print("Score = {0:.5f}".format(score))
```

```python
# 점수가 다른 이유 : 할때 마다 달라진다. 의사결정 트리가 랜덤하게 처리하므로
# 랜덤하게 나오지 않도록 고정시켜 주어야 한다.  
# model = DecisionTreeRegressor(random_state=37)
1.25369
1.22381
```

```python
#학습
model.fit(X_train, y_train)
model
```

```python
predictions = model.predict(X_test)
```
