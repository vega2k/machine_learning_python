### barplot

```python
sns.barplot(data=train, x="Pclass", y="Fare", hue="Survived")
```

#countplot은 x,y 를 동시에 지정할 수 없다. ,x 나 y 둘중에 하나씩 만 쓸 수 있다.  classification 은 countplot
#barplot x,y 둘다 지정가능, y축에 숫자를 지정할수 있는데 평균값으로  regression 은 barplot 에서 주로 사용
### bar plot은 y(정수형)를 지정하면, y의 평균값을 보여준다. 
#### (x,y)를 바꿀수도 있지만 기본적으로 y는 정수형이어야 한다. 가운데의 직선은, 표준편차를 의미한다.
### 풀려는 문제가 분류문제면 count plot을 사용하고, 회귀 문제면 bar plot을 사용할 수 있다. 

### pointplot

### x 축이 서로 연관성이 있을 경우에는 point plot
### x 축이 서로 연관성이 없을 경우에는 bar plot
### bar plot은 연관관계가 잘 보이지 않는다. 연관관계를 보다 잘 확인하기 위해, point plot으로 바꿔보자 :
### 꺾은선 그래프(point plot)는 연관관계를 보려할 때 쓴다. 운임요금은 객실등급이 높을수록 대폭 상승하는 걸 알 수 있다. 
### 이렇게 했을때, '돈을 많이 내면 살 확률이 높다'는 가정은 어느정도 타당하다는 결론에 도달할 수 있다. 


```python
sns.pointplot(data=train, x="Pclass", y="Fare", hue="Survived")
```

### distPlot

### 객실 등급과 다르게, 나이는 너무나도 값이 다양하다. 이런 상황에서는 어떻게 시각화하는 것이 좋을까? 
### dist plot을 이용하면 분포도를 확인할 수 있다.
### ※ Age에는 빈 값이 들어있다. 빈 값은 아래처럼 라이브러리 오류를 발생시킬 수 있기 때문에 주의해야 한다 :

```python
sns.distplot(train["Age"],hist=False)
```

```python
mean_age = train['Age'].mean()
train.loc[pd.isnull(train["Age"]), "Age"] = mean_age
train[pd.isnull(train["Age"])]
```

```python
sns.distplot(train["Age"],hist=False)
```

```python
#전체적으로 어떻게 분포되어 있는지를 볼 수 있다. 
#sns.distplot(data=train, x="Fare)
sns.distplot(train["Fare"])
```

```python
#가로축을 bin 이라고 함

sns.distplot(train["Fare"], hist=False)
```

```python
#500 달러 낸  사람은 outlier
train['Fare']
low_fare = train[train['Fare'] < 100]
sns.distplot(low_fare["Fare"], hist=False)
```

```python
#너무 낮은 운임요금을 지불한 승객들은 죽을 확률이 매우 높았다.
perish = low_fare[low_fare["Survived"] == 0]
survived = low_fare[low_fare["Survived"] == 1]

sns.distplot(perish["Fare"], hist=False)
sns.distplot(survived["Fare"], hist=False)
```

```python
sns.distplot(perish["Fare"], hist=False, label="Perish")
sns.distplot(survived["Fare"], hist=False, label="Survived")
```

### lmplot

```python
sns.lmplot(data=train, x="Age",y="Fare",hue="Survived")
```

```python
#회귀선 사라지게 하는 옵션
sns.lmplot(data=train, x="Age",y="Fare",hue="Survived",fit_reg=False)
```

```python
low_fare = train[train["Fare"] < 500]
sns.lmplot(data=low_fare, x="Age",y="Fare",hue="Survived",fit_reg=False)
```

```python
low_fare = train[train["Fare"] < 100]
sns.lmplot(data=low_fare, x="Age",y="Fare",hue="Survived",fit_reg=False)
```
